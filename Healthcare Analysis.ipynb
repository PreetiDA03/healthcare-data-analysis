```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

ax = sns.countplot(x='Medical_Condition', data=df, hue='Medical_Condition', palette='viridis', legend=False)

plt.title("Number of Patients by Medical Condition")
plt.xlabel("Medical Condition")
plt.ylabel("Number of Patients")

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()


for p in ax.patches:
    ax.annotate(int(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

## Patient distribution across medical conditions is fairly balanced, with no single condition dominating significantly.

```python
import calendar

plt.figure(figsize=(10,6))
df['Month_Name'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df.groupby('Month_Name').size().plot(marker='o')
plt.title("Monthly Patient Admissions Trend")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Month")
plt.ylabel("Number of Patients")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Patient admissions remain relatively stable throughout the year, indicating consistent healthcare demand without strong seasonal variation.

```python
plt.figure(figsize=(10,6))
df.groupby('Medical_Condition')['Billing_Amount'].mean().sort_values().plot(kind='bar')
plt.title("Average Billing Amount by Medical Condition")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Medical Condition")
plt.ylabel("Average Billing Amount")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Certain medical conditions have higher average billing, suggesting that treatment complexity directly impacts healthcare costs.

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Admission_Type', y='Billing_Amount', data=df, palette='Set2')
plt.title("Billing Amount Distribution by Admission Type")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Admission Type")
plt.ylabel("Billing Amount")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Emergency admissions tend to have higher billing amounts compared to elective and urgent cases, indicating higher treatment intensity.

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Gender', y='Billing_Amount', data=df, palette='coolwarm')
plt.title("Billing Amount Comparison by Gender")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Gender")
plt.ylabel("Billing Amount")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Billing amounts are similar across genders, indicating no significant cost variation between male and female patients.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Stay_Days'], bins=30, kde=True)
plt.title("Distribution of Patient Stay Duration")
plt.xticks(rotation=45, ha='right')
plt.xlabel("Stay Days")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Most patients have a moderate hospital stay duration, with fewer cases of extremely short or long stays.
